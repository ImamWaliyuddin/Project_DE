```python
df=pd.DataFrame(columns=["customer_id","credit_score","country","gender","age","tenure","balance","products_number","credit_card","active_member","estimated_salary"])
for i in data:
    df_tmp={"customer_id":i[0],"credit_score":i[1],"country":i[2],"gender":i[3],"age":i[4],"tenure":i[5],"balance":i[6],"products_number":i[7],"credit_card":i[8],"active_member":i[9],"estimated_salary":i[10]}
    df=df.append(df_tmp, ignore_index = True)
df.head(10)
```
